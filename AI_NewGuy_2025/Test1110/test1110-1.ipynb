```python
# 有一個直角三角形，已知兩邊長，求第三邊長
# 已知兩邊長，求第三邊長

import math#import math引入數學工具箱 

adjacent_side = 3  # 鄰邊長度
opposite_side = 4  # 對邊長度

# 使用畢氏定理計算斜邊長度
hypotenuse_length = math.sqrt(adjacent_side**2 + opposite_side**2)
#math.sqrt() 這個函式來計算平方根
print(f"斜邊長度 = {hypotenuse_length:.2f}")

#f，指的是 f-string,f-string 的用途： 讓你在字串（文字）中可以直接嵌入變數的值或表達式的計算結果。
#花括號 {}它告訴 Python：「把 {} 裡面的東西當作 Python 程式碼來執行，然後把結果變成文字放進來。」
#.2f 的解釋：.：小數點。2：保留兩位小數。f：表示這個數值是一個浮點數（帶有小數點的數字，float）。
```

```python
i = 10
#複合指定運算子
i += 50
i
# += 是複合指定運算子，意思是把 i 的值加上 50 後，再存回 i 變數。
```

```python
i = 0
for j in range(1,11):
    i += j
i
#for j in range(1,11): 這行是說，從 1 開始，到 11 結束，每次增加 1。
#for 迴圈 (Loop) for 就像一個點名機或自動重複機。當你需要**「針對一組東西，一個接一個地執行同樣的動作」
```

```python
# for 這是一組要處理的「東西」
students = ["小明", "小華", "小美"]

# 迴圈開始：
# 對於清單中的「每一個」項目（我們命名為 student），執行冒號後縮排的程式碼
for student in students:
    print(f"你好，{student}！") # 這裡使用了剛剛學過的 f-string
```
